```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data_extracted = pd.read_csv('C:\\Users\\Bilal\\Desktop\\Extracted_Data.csv')
```

```python
# displaying data
data_extracted
```

```python
# Year-by-Year growth rating for Total Revenue and Net Income
data_extracted['Revenue Growth by (%)'] = data_extracted.groupby('Company')['Total Revenue'].pct_change() * 100
data_extracted['Net Income Growth by (%)'] = data_extracted.groupby('Company')['Net Income'].pct_change() * 100
data_extracted
```

```python
# Imputing Null/NaN values that result from pct_change calculations with 0

data_extracted.fillna(0, inplace=True)
```

```python
data_extracted
```

```python
#  Year-by-Year growth rates for Total Assets, Total Liabilities and Cash flow from Ops Activities
data_extracted['Assets Growth (%)'] = data_extracted.groupby('Company')['Total Assets'].pct_change() * 100
data_extracted['Liabilities Growth (%)'] = data_extracted.groupby('Company')['Total Liabilities'].pct_change() * 100
data_extracted['Cash Flow from Operations Growth(%)'] = data_extracted.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
data_extracted
```

```python
data_extracted.fillna(0, inplace=True)
data_extracted
```

```python
# Saving the CSV
data_extracted.to_csv('C:\\Users\\Bilal\\Desktop\\Finaldf.csv')
```

```python
summary_byCompany = data_extracted.groupby('Company').agg({
    'Revenue Growth by (%)': 'mean',
    'Net Income Growth by (%)': 'mean',
    'Assets Growth (%)' : 'mean',
    'Liabilities Growth (%)' : 'mean',
    'Cash Flow from Operations Growth(%)' :'mean'    
}).reset_index()
```
